```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df), columns= df.columns)
scaled_df.head(10)
```

```python
features = scaled_df.drop(columns=['motor_speed'])
target = scaled_df['motor_speed']
features.head(5)
```

```python
target.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state= 100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Using Linear Regression Model
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
prediction =pd.DataFrame( linear_model.predict(x_test), columns=['linear_pred'])
prediction
```

```python
#Finding performance matrix
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
linear_mae = mean_absolute_error(y_test, prediction['linear_pred'])
print('Mean absolute error', linear_mae)
linear_mse  = mean_squared_error(y_test,prediction['linear_pred'])
print('Mean squared error', linear_mse)
linear_r2 = r2_score(y_test,prediction['linear_pred'])
print('r2 score', linear_r2)
```

```python
#Using K Nearest Naighbour Model
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(x_train, y_train)
```

```python
prediction['knn_pred'] = knn_model.predict(x_test)
prediction
```

```python
#Applying Performance matrix on KNN
```

```python
knn_mae = mean_absolute_error(y_test, prediction['knn_pred'])
print('Mean absolute error', knn_mae)
knn_mse  = mean_squared_error(y_test, prediction['knn_pred'])
print('Mean squared error', knn_mse)
knn_r2 = r2_score(y_test, prediction['knn_pred'])
print('r2 score', knn_r2)
```

```python
#Using Random Forest Model
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=10, random_state=42)
rf_model.fit(x_train, y_train)
```

```python
prediction['rf_pred'] = rf_model.predict(x_test)
prediction
```

```python
#Applying Perfomance matrix for Random Forest
```

```python
rf_mae = mean_absolute_error(y_test, prediction['rf_pred'])
print('Mean absolute error', rf_mae)
rf_mse  = mean_squared_error(y_test, prediction['rf_pred'])
print('Mean squared error', rf_mse)
rf_r2 = r2_score(y_test, prediction['rf_pred'])
print('r2 score', rf_r2)
```

```python
#Report of 3 models
```

```python
report = pd.DataFrame({'Models':['LinearRegression Model','KNeighborsRegressor Model','RandomForestRegressor'], 'Mean_adsolute_error':[linear_mae, knn_mae, rf_mae],'Mean_squared_error':[linear_mse, knn_mse, rf_mse], 'R2_error':[linear_r2, knn_r2, rf_r2]})
report
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x=report['Mean_adsolute_error'], 
                y=report['R2_error'], 
                hue=report['Models'])
```

```python
#we are getting less errors in Random Forest Model
#We are finding Random Forest Model as Good Model
```
